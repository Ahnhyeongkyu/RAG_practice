## 통합형 Loader 인터페이스

- loader 객체: 다양한 로더 도구를 활용하여 loader 객체를 생성합니다.
- 생성된 loader 객체의 load() 함수는 전체 문서를 로드하여 반환합니다.
- 생성된 loader 객체의 load_and_split() 함수는 문서를 split 한 결과를 반환합니다.
- 반환된 document 구조는 page_content 와 metadata 속성값을 포함합니다. page_content 는 문서의 내용을, metadata 는 파일명, 페이지 번호, 그 밖에 유용한 메타정보를 포함합니다.

```python
import subprocess
import json
from pprint import pprint

class JSONLoader:
    def __init__(self, file_path, jq_schema, text_content=True):
        self.file_path = file_path
        self.jq_schema = jq_schema
        self.text_content = text_content

    def load(self):
        # jq 명령 실행
        result = subprocess.run(
            ['jq', self.jq_schema, self.file_path],
            capture_output=True,
            text=True
        )
        
        # 결과 파싱
        if result.returncode != 0:
            raise ValueError(f"Error executing jq: {result.stderr}")

        # jq 출력 디버깅
        print("jq command output:")
        print(result.stdout)
        
        try:
            # JSON 배열로 감싸서 파싱
            data = json.loads(f"[{result.stdout.strip()}]")
        except json.JSONDecodeError as e:
            print("Failed to decode JSON output from jq command.")
            print("jq output:", result.stdout)
            raise e
        
        if self.text_content:
            return data
        else:
            return [{"phoneNumbers": item} for item in data]

# JSON 파일 경로
json_file_path = "data/people.json"

# JSONLoader 인스턴스 생성
loader = JSONLoader(
    file_path=json_file_path,
    jq_schema=".[].phoneNumbers",
    text_content=False
)

# 데이터 로드
data = loader.load()

# 데이터 출력
pprint(data)
```

### TextLoader를 통한 파일 인코딩 자동 감지

- silent_errors: 디렉토리로더에 silent_errors 매개변수를 전달하여 로드할 수 없는 파일을 건너뛰고 로드 프로세스를 계속할 수 있습니다.
- autodetect_encoding: 또한 로더 클래스에 자동 감지_인코딩을 전달하여 실패하기 전에 파일 인코딩을 자동으로 감지하도록 요청할 수도 있습니다.

```python
path = "data/"

text_loader_kwargs = {"autodetect_encoding": True}

loader = DirectoryLoader(
    path,
    glob="**/*.txt",
    loader_cls=TextLoader,
    silent_errors=True,
    loader_kwargs=text_loader_kwargs,
)
docs = loader.load()
```

```python
doc_sources = [doc.metadata["source"] for doc in docs]
doc_sources
```
